```python
# IGNORE THIS FOR NOW - ATTEMPTING TO CALCULATE THE DISTANCE OF SOLAR ORBITER FROM THE SUN USING SUNPY HORIZONS MODULE
# This is taking ages, for example 650,000 time points takes more than 15 minutes to query, and we have 20 times more than that
from sunpy.coordinates import frames
from sunpy.coordinates import get_horizons_coord
from astropy.coordinates import SkyCoord
from astropy import units as u
R_SUN_M = (u.R_sun.to(u.m))

def split_list_into_chunks(input_list, chunk_size=10000):
    """
    Splits a list into smaller lists (chunks) of a specified size.

    Args:
        input_list (list): The list to be split.
        chunk_size (int): The maximum size of each chunk. Defaults to 10000.

    Returns:
        list: A list of sub-lists (chunks).
    """
    # Use a list comprehension with slicing to generate the chunks
    return [
        input_list[i:i + chunk_size]
        for i in range(0, len(input_list), chunk_size)
    ]

def get_solo_ephemeris(times) -> pd.DataFrame:
    """Fetch Solar Orbiter positions from JPL Horizons via SunPy and convert to HGC.

    Returns DataFrame with columns: time (datetime64[ns, UTC]),
    lon_sc_deg, lat_sc_deg, r_sc_Rsun
    """
    print(f"Querying Horizons for Solar Orbiter coords at N={len(times)} timesâ€¦")
    all_dfs = []


    for time_chunk in split_list_into_chunks(times, chunk_size=10000-1):
        # repeat for each chunk
        sc = get_horizons_coord("Solar Orbiter", time_chunk)
        # Convert to HeliographicCarrington (Carrington long/lat)
        hgc = sc.transform_to(frames.HeliographicCarrington(observer="sun"))

        # Extract longitude/latitude/radius
        lon = hgc.lon.to(u.deg).value  # 0..360
        lat = hgc.lat.to(u.deg).value
        r_m = hgc.radius.to(u.m).value
        r_Rsun = r_m / R_SUN_M

        df = pd.DataFrame({
            "time": time_chunk,
            "lon_sc_deg": lon,
            "lat_sc_deg": lat,
            "r_sc_Rsun": r_Rsun,
        })
        all_dfs.append(df)
    
    # Finally concatenate all chunked timeseries dataframes
    df = pd.concat(all_dfs)
    return df

# not complete yet - need to join the ephemeris data to the timeseries dataframes
# swa_ephemeris = get_solo_ephemeris(SWA_timeseries_df.index)
# SWA_timeseries_df = pd.merge( SWA_timeseries_df, swa_ephemeris, how="left", left_index=True, right_on='time')

# mag_ephemeris = get_solo_ephemeris(MAG_timeseries_df.index)
# MAG_timeseries_df = pd.merge( MAG_timeseries_df, mag_ephemeris, how="left", left_index=True, right_on='time')
```
